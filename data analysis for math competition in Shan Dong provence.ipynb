# 2018年全国大学生数学竞赛山东省获奖统计分析

```python
# 加载文件
import pandas as pd
file_name1 = './shengjiang.xlsx'
data_sheng = pd.read_excel(file_name1,encoding = 'utf-8')
data_sheng.head()
```

```python
# 删掉不需要的列
data_new = data_sheng.drop(columns=[u'身份证号',u'学院'])
data_new.head()
```

```python
# 查看数据的纬度
print(data_new.shape)

# 查看有什么学校参加
school = data_new[u'学校'].unique()
print(school)
print('-----一共有 {} 所学校参加-----'.format(len(school)))
```
